```python
from bs4 import BeautifulSoup
import urllib.request
from urllib.error import HTTPError
from urllib.parse import urlparse
from urllib.parse import urljoin
from time import sleep

result_set = set()

# a
def parse_response(response):
        soup = BeautifulSoup(response.read(), 'html.parser')
        for a in soup.find_all("a"):
            yield a

def do_request(url):
        
        # 
        if url in result_set:
            return
        
        request_target = urlparse(url)
        if request_target.hostname != 'xxxxx.xxx.xxxx' and request_target.hostname != 'xxx.xxx.xxx':
            return

        try:
            # URL
            result_set.add(url)
            # 
            sleep(1)
            response = urllib.request.urlopen(url)
        except HTTPError as e:
            # 200URL
            print(str(e.code) + ":" + url)
            return
                
        for tag in parse_response(response):
            if not tag.has_attr('href'):
                continue
            href = tag['href']
            target = urlparse(href)
            if target.scheme == 'javascript':
                #javascript()
                #print('javascript')
                continue
            elif target.scheme == 'http' and target.scheme == 'https':
                target = urlparse(href)
                do_request(target)
            else:
                joined_url = urljoin(url, href)
                do_request(joined_url)

print("======================= start =======================")                
do_request("https://xxxxxxxx.xxx.xxx/")
print("=======================finish =======================")
```
